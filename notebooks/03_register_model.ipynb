```python
import mlflow
from mlflow.tracking import MlflowClient

mlflow.set_tracking_uri("http://127.0.0.1:5000")
client = MlflowClient()
model_name = "bank_churn_model"
```

```python
#search for the run programmatically
experiment = mlflow.get_experiment_by_name("bank-churn-experiment")
print("Experiment ID:", experiment.experiment_id)
```

```python
#get the runs
runs = mlflow.search_runs(
    experiment_ids=[experiment.experiment_id],
    order_by=["start_time DESC"]
)

runs[["run_id", "tags.mlflow.runName", "metrics.accuracy", "metrics.f1"]]
```

```python
#get Random Forest ID:
rf_run = runs[runs["tags.mlflow.runName"] == "random_forest"].iloc[0]
run_id = rf_run["run_id"]
print("Random Forest run_id:", run_id)
```

```python
#register using that run id
model_uri = f"runs:/{run_id}/random_forest_model"

registered_model = mlflow.register_model(
    model_uri=model_uri,
    name=model_name
)

print("Model registered:")
print("Name:", registered_model.name)
print("Version:", registered_model.version)
```
